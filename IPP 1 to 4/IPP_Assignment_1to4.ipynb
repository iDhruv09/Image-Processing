### Que: 7 WAP that roatates an image by 90 degrees clockwise using OpenCV

```python
(h, w) = img.shape[:2]
```

```python
center = (w // 2, h // 2)
```

```python
M = cv2.getRotationMatrix2D(center, -90, 1.0)
```

```python
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated 90 Clockwise")
plt.axis("off")
plt.show()
```

### Que: 8 WAP statement to flip the image "car.jpg" left to right using OpenCV and display result.

```python
car = cv2.imread("car.jpg")
```

```python
plt.imshow(cv2.cvtColor(car,cv2.COLOR_BGR2RGB))
plt.title("Orignal Image")
plt.axis('off')
plt.show()
```

```python
# Answer:
flipped = cv2.flip(car, 1)
```

```python
plt.imshow(cv2.cvtColor(flipped,cv2.COLOR_BGR2RGB))
plt.title("Flipped Image")
plt.axis('off')
plt.show()
```

```python
cv2.imwrite('flipped_car.jpg',flipped)
```
